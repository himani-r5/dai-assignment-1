# DATA CLEANING

```python
data["Educational_level"]=data["Educational_level"].fillna("None")
```

```python
data.drop('Defect_of_vehicle',axis=1,inplace=True)
```

```python
data['Service_year_of_vehicle']=data['Service_year_of_vehicle'].fillna(data['Service_year_of_vehicle'].mode()[0])
```

```python
data["Work_of_casuality"].fillna("Unknown",inplace=True)
```

```python
data['Type_of_vehicle']=data['Type_of_vehicle'].fillna('unknown')
```

```python
data['Type_of_vehicle']=data['Type_of_vehicle'].replace('Public (13?45 seats)','Public (13-45 seats)')
data['Type_of_vehicle']=data['Type_of_vehicle'].replace('Public (> 45 seats)','Public (more than 45 seats)')
data['Type_of_vehicle']=data['Type_of_vehicle'].replace('Lorry (11?40Q)','Lorry')
data['Type_of_vehicle']=data['Type_of_vehicle'].replace('Lorry (41?100Q)','Lorry')
data['Type_of_vehicle']=data['Type_of_vehicle'].replace('Pick up upto 10Q','Pick up')
```

```python
vm_mode=data['Vehicle_movement'].mode()
data['Vehicle_movement']=data['Vehicle_movement'].fillna(vm_mode.values[0])
data['Vehicle_movement'].isnull().sum()
```

```python
de_mode=data['Driving_experience'].mode()
data['Driving_experience']=data['Driving_experience'].fillna(de_mode.values[0])
data['Driving_experience'].isnull().sum()
```

```python
data["Sex_of_driver"]=data["Sex_of_driver"].replace("Unknown","Other")
```

```python
data.drop("Fitness_of_casuality",axis=1,inplace=True)
```

```python
data['Time']=data['Time'].astype('datetime64[ns]')
data.loc[::,"Time"].reset_index()
```

```python
data.duplicated().sum()
```

```python
data.dropna(inplace=True)
```

## Vehicle Driver demographics

```python
plt.figure(figsize=(15,10))
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.5, 
                    hspace=0.8)
plt.subplot(2,2,1)
data["Age_band_of_driver"].value_counts().plot(kind="bar")
plt.xlabel("Age group")
plt.ylabel("Number of accidents")
plt.title("Age group distribution")

plt.subplot(2,2,2)
data["Educational_level"].value_counts().plot(kind="bar")
plt.xlabel("Educational level")
plt.ylabel("Number of accidents")
plt.title("Education level distribution")

plt.subplot(2,2,3)
data["Sex_of_driver"].value_counts().plot.pie(autopct='%1.1f%%')
plt.xlabel('')
plt.ylabel('')
plt.title("Gender distribution")

plt.subplot(2,2,4)
data["Driving_experience"].value_counts().plot(kind="bar")
plt.xlabel("Driving experience")
plt.ylabel("Number of accidents")
plt.title("Driving experience distribution")

plt.suptitle("Vehicle Driver demographics")
plt.show()
```

# Bivariate Analysis  

```python
sev_count=data.groupby('Day_of_week')['Accident_severity'].value_counts().unstack()
sev_count.plot(kind='bar',stacked=True)
plt.title("Accident Severity by day of week")
plt.xlabel('Days of week')
plt.ylabel('Number of accidents')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Accident_severity', y='Hour', data=data , palette='coolwarm')
plt.title('Hour vs Accident Severity')
plt.xlabel('Accident Severity')
plt.ylabel('Hour of the Day')
plt.show()
```

```python
sns.countplot(x='Age_band_of_driver',hue='Accident_severity',data=data)
plt.title("Accident Severity by age of driver")
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
find_cor = data[['Number_of_vehicles_involved', 'Number_of_casualties']]
correlation_mat = find_cor.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(data=correlation_mat, annot=True, cmap='YlGnBu', fmt='.2f', linewidths=1)
plt.title('Correlation Plot: Number of Vehicles Involved vs. Number of Casualties')
plt.show()
```

```python
accident_cause_movement = data.groupby(['Vehicle_movement', 'Cause_of_accident']).size().reset_index(name='Count')

straight_movement_data = accident_cause_movement[accident_cause_movement['Vehicle_movement'] == 'Going straight']
plt.figure(figsize=(10,6))
sns.barplot(x='Cause_of_accident', y='Count', data=straight_movement_data, palette='Blues_d')
plt.title('Accidents Caused While Moving Straight ')
plt.xlabel('Cause of Accident')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=90)
plt.show()
```

# INSIGHTS 

### Vehicle Demographics
- Most accidents involve drivers in the age group of 18-30 showing lack of experience
- A large proportion of drivers do not have formal education ( only till primary ), showing lack of road safety knowledge
- Male drivers are more involved in accidents than female drivers
- Drivers with limited experience(1-5 years) are more involved in accidents showing that Driving experience is a major contributor to accidents
- Majorly , automobiles are involved in accidents

### Accident Demographics
- Evenly distributed throughout the week , more prone to happen near weekends(Thurs-Friday)
- Higher concerntration of accidents in peak traffic hours (8 AM-6 PM)
- Y junction has maximum density of accidents happening , so one must be careful while crossing these
- Majority of accidents have severity has "Slight" compared to the lesser proportion of Serious ones

### Numerical Demographics
- Majority number of casualty is 1 which indicates the casualty role being the Driver
- Number of vehicles involved are mostly 2 , resulting in lesser number of casualty 
- Correlation between number of vehicles involved and number of casualty is 0.22


### Bivariate Analysis
- Accident severity increases(Fatal) on weekends
- Younger drivers (18-30 years) tend to have more severe accidents compared to older drivers.
- The heatmap shows a positive correlation between the number of vehicles involved and the number of casualties, indicating that multi-vehicle collisions tend to be more dangerous.
